```python
import pickle
import pandas as pd
from sklearn import svm
from UtilityFunctions import *
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
Labels = ['country', 'id', 'title', 'c1', 'c2', 'c3', 'description', 'price', 'type']
```

```python
df = pd.read_csv('data_train.csv',names = Labels)
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf, Y1, Y2, Y3 = cleaning_data(df, tfidf_vectorizer)
```

# Training Models

<!--  -->

## For Category # 1

```python
x_train, x_test, y_train, y_test = train_test_split(X_train_tfidf, Y1, test_size=0.3, random_state=42)
clf_svm_cat_one = svm.LinearSVC().fit(x_train, y_train)
```

```python
y_pred_cat_one = clf_svm_cat_one.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_cat_one)
```

```python
print(classification_report(y_test, y_pred_cat_one))
```

```python
mat = confusion_matrix(y_test, y_pred_cat_one)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(mat.T, square=False, annot=True, fmt='d', cbar=True
            , xticklabels= Y1.unique(), yticklabels= Y1.unique())

plt.ylabel('Actual lable')
plt.xlabel('Predicted lable')
```
